```python
import mcp
```

```python
mod = dir(mcp)
print("MCP module contents:")
for item in mod:
    if not item.startswith("_"):
        print(item)
print("\nMCP module functions:")
for item in mod:
    if item.startswith("_"):
        continue
    func = getattr(mcp, item)
    if callable(func):
        print(f"{item}() - {func.__doc__}")
print("\nMCP module classes:")
for item in mod:
    if item.startswith("_"):
        continue
    cls = getattr(mcp, item)
    if isinstance(cls, type):
        print(f"{item} - {cls.__doc__}")

print("\nMCP module attributes:")
for item in mod:
    if item.startswith("_"):
        continue
    attr = getattr(mcp, item)
    if not callable(attr) and not isinstance(attr, (int, float, str)):
        print(f"{item} - {attr}")


```
